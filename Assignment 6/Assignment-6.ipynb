```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

Data=pd.read_csv("weather.csv")
Data.head()
```

```python
print(Data.columns.tolist())
Data.describe(include="all")
```

```python
sns.scatterplot(data=Data,x="Sunshine",y="Rainfall")
plt.show()
```

```python
# sunshine in regards to clouds in sky at 9am and 3pm
sns.scatterplot(data=Data,x="Cloud9am",y="Sunshine")
plt.show()
```

```python
sns.scatterplot(data=Data,x="Cloud3pm",y="Sunshine")
plt.show()
```

```python
#Does Temprature Depend on WindSpeed?
sns.scatterplot(data=Data,x="WindGustSpeed",y='MinTemp',hue='Sunshine')
plt.show()
```

```python
sns.scatterplot(data=Data,x="WindGustSpeed",y='MaxTemp')
plt.show()
```

```python
sns.scatterplot(data=Data,y="Evaporation",x="Sunshine")
plt.show()
```

```python
# how sunshine effects rainchance
sns.swarmplot(data=Data,y="Sunshine",x="RainToday")
plt.show()
```

```python
# Does Clouds at 9am mean it will Rain Today?
sns.swarmplot(data=Data,y="Cloud9am",x="RainToday")
plt.show()
```

```python
# Does Clouds at 3pm mean it will Rain Tomoorrow?
sns.swarmplot(data=Data,y="Cloud3pm",x="RainTomorrow")
plt.show()
```

```python
# How much does Temp at 9am mean it will Rain Today?
sns.swarmplot(data=Data,y="Temp9am",x="RainToday")
plt.show()
```

```python
# How much does Temp at 3pm mean it will Rain Tomorrow?
sns.swarmplot(data=Data,y="Temp3pm",x="RainTomorrow")
plt.show()
```

```python
sns.kdeplot(data=Data['Cloud9am'],data2=Data["Temp9am"],shade=True)
plt.show()
sns.kdeplot(data=Data['Cloud9am'],data2=Data["Temp9am"],shade=True)
plt.show()
```

```python
sns.kdeplot(data=Data['Cloud3pm'],data2=Data["Temp3pm"],shade=True)
plt.show()
sns.kdeplot(data=Data['Cloud3pm'],data2=Data["Temp3pm"],shade=True)
plt.show()
```

```python
sns.distplot(Data['Evaporation'])
plt.show()
```

```python
sns.boxplot(data=Data,y="Sunshine",x="RainToday")
plt.show()
```

```python
sns.boxplot(data=Data,y="Sunshine",x="RainTomorrow")
plt.show()
```
