```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Data = pd.read_csv('House_Price.csv')
print(Data.describe(include='all'))
```

```python
Data.head()
```

```python
sns.scatterplot(Data['bedrooms'],Data['price'])
plt.show()
```

```python
sns.boxplot(y=Data['bedrooms'])
plt.show()
```

Above it shows there might be some outliers in the data, so I ommit any house with rooms > 9.

```python
sns.boxplot(y=Data['bathrooms'])
```

Also above data shows that there might be some data outliers for bathrooms. So I remove house data with bathrooms >5

```python
Data=Data[Data['bedrooms']<10]
Data=Data[Data['bathrooms']<6]
```

```python
sns.scatterplot(Data['bedrooms'],Data['price'])
plt.show()
```

```python
sns.scatterplot(Data['bathrooms'],Data['price'])
plt.show()
```

```python
sns.scatterplot(Data['sqft_living'],Data['price'])
plt.show()
```

We use heat map to check corelation values across the data.

```python
Corr=Data.corr()
plt.figure(figsize=(15,15))
sns.heatmap(Corr,annot=True,vmax=1,vmin=-1)
plt.show()
```
